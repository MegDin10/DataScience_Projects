## Importing the Necesasary Libraries 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Prices

```python
fig , ax = plt.subplots(1,2,figsize = (20,5))
sns.distplot(df[df.neighbourhood_group == "Manhattan"].price,ax = ax[0] )
sns.distplot(df[df.neighbourhood_group == "Brooklyn"].price,ax = ax[1])

fig , ax = plt.subplots(1,2,figsize = (20,5))
sns.distplot(df[df.neighbourhood_group == "Staten Island"].price,ax = ax[0] )
sns.distplot(df[df.neighbourhood_group == "Bronx"].price,ax = ax[1] )

fig = plt.subplots(figsize = (7.5,4))
sns.distplot(df[df.neighbourhood_group == "Queens"].price )

print("Mean Price to stay in Manhattan  is :",int(df[df.neighbourhood_group == "Manhattan"].price.mean()))
print("Mean Price to stay in Brooklyn  is :",int(df[df.neighbourhood_group == "Brooklyn"].price.mean()))

print("Mean Price of in Staten Island is : ", int(df[df.neighbourhood_group == "Staten Island"].price.mean()))
print("Mean Price of in Bronx is : ", int(df[df.neighbourhood_group == "Bronx"].price.mean()))
print("Mean Price of in Queens is : ", int(df[df.neighbourhood_group == "Queens"].price.mean()))
```

* We see Through the above Graphs and Mean values how Price varies According to the Main Neigbhourhood Groups.
* Now lets first try to see what  it would take to stay in New York on a Budget.

```python
f, ax = plt.subplots(figsize = (10,5))
sns.countplot(x = df.room_type , data = df)
df.groupby(['room_type'])['id'].count()/len(df.room_type)*100
```

* Since no of shared rooms , are Significantly less , there is no harm in saying that people who normally use Airbnb service to visit New York City, are probably NOT that keen on saving. 
* If they were there would have been an apreciatable number of Shared rooms.
* People on a tight budget try to book before hand or else you wont find a place to stay .


```python
plt.figure(figsize = (12,5))
Mean_Price = df.groupby(['room_type',"neighbourhood_group"])['price'].mean().sort_values(ascending = False)
Mean_Price.plot(kind = 'bar')
plt.title('Average Price for Different Types rooms in neighbourhood group')
plt.ylabel('Average Daily Price')
plt.xlabel('Neighbourhood Group')
plt.show()
pd.DataFrame(Mean_Price)
```
